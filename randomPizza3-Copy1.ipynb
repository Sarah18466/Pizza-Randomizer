```python
import mysql.connector
mydb = mysql.connector.connect(
    host="db_host",
    user="db_user",
    password="db_password",
    db="db_name")
mycursor = mydb.cursor()
import random
```

```python
class new_person:
    def __init__(self):
        self.name = input("Enter username:")
    
    def __str__(self):
        return f"{self.name}"

    def sign_up(self):
        find = "SELECT COUNT(personid) FROM person WHERE username='"+self.name+"';"
        mycursor.execute(find)
        cursor = []
        for x in mycursor:
            cursor.append(x)
        c = list(cursor)
        c = str(c[0])
        c = c[1]
        c = int(c)
        if c == 0:
            password = input("Enter password:")
            password = str(password)
            insert = "INSERT INTO person (username, password) VALUES ('"+self.name+"', '"+password+"');"
            mycursor.execute(insert)
            mydb.commit()
            self.get_allergy()
            self.get_group()
            self.get_type()
            print("Welcome to the pizza randomizer! Login to start!")
        else:
            print("username already taken")
            return new_person()

    def get_allergy(self):
        print("Do you have any allergies? Options are milk, egg, wheat and/or soy. Capitalize each separate word with a space in between. If there are none, hit enter.")
        x = input()
        allergy = str.split(x)
        nextAllergy=[]
        for i in range(len(str.split(x))):
            nextAllergy.append("INSERT INTO customer VALUES ('"+self.name+"', "+"'allergen', '"+allergy[i]+"');")
        for i in range(len(str.split(x))):
            mycursor.execute(nextAllergy[i])
            mydb.commit()
    
    def get_group(self):
        print("Are there any food groups you would like to delete? Ex) meat, vegetables, and/or fruit. If there are multiple, capitalize each one. If there are none, hit enter.")
        y = input()
        group = str.split(y)
        nextGroup=[]
        for i in range(len(str.split(y))):
            nextGroup.append("INSERT INTO customer VALUES ('"+self.name+"', "+"'foodGroup', '"+group[i]+"');")
        for i in range(len(str.split(y))):
            mycursor.execute(nextGroup[i])
            mydb.commit()
        
    def get_type(self):
        print("Are there any food types you would like to delete? Ex) pork, beef, chicken. If there are multiple, capitalize each one. If there are none, hit enter.")
        z = input()
        foodType = str.split(z)
        nextType=[]
        for i in range(len(str.split(z))):
            nextType.append("INSERT INTO customer VALUES ('"+self.name+"', "+"'foodType', '"+foodType[i]+"');")
        for i in range(len(str.split(z))):
            mycursor.execute(nextType[i])
            mydb.commit()
```

```python
class old_person:
    def __init__(self):
        self.name = input("Enter username:")
        self.password = input("Enter password:")
    
    def __str__(self):
        return f"{self.name}"
    
    def login(self):
        find = "SELECT COUNT(personid) FROM person WHERE username='"+self.name+"' AND password='"+self.password+"';"
        mycursor.execute(find)
        cursor = []
        for x in mycursor:
            cursor.append(x)
        c = list(cursor)
        c = str(c[0])
        c = c[1]
        c = int(c)
        if c == 1:
            print("Welcome back!")
            self.randomizer()
        else:
            print("Wrong")
            return old_person()
        
    def randomizer(self):
        new = "SELECT foodName FROM topping WHERE NOT EXISTS (SELECT abstain FROM customer WHERE (topping.allergen=customer.abstain OR topping.foodType=customer.abstain OR topping.foodGroup=customer.abstain) AND username='"+self.name+"');"
        cursor=[]
        mycursor.execute(new)
        for x in mycursor:
            cursor.append(x)
        print("Number of toppings available:", len(cursor))
        print(cursor)
        print()
        f = len(cursor)

        sauce = "SELECT sauceName FROM sauce WHERE NOT EXISTS (SELECT abstain FROM customer WHERE sauce.allergen=customer.abstain AND username='"+self.name+"');"
        s_cursor=[]
        mycursor.execute(sauce)
        for x in mycursor:
            s_cursor.append(x)
        print("Number of sauces available:", len(s_cursor))
        print(s_cursor)
        print()

        crust = "SELECT crustName FROM crust WHERE NOT EXISTS (SELECT abstain FROM customer WHERE crust.allergen=customer.abstain AND username='"+self.name+"');"
        c_cursor=[]
        mycursor.execute(crust)
        for x in mycursor:
            c_cursor.append(x)
        print("Number of crusts available:", len(c_cursor))
        print(c_cursor)
        print()
        n = get_amount(f)
        random.shuffle(cursor)
        random.shuffle(s_cursor)
        random.shuffle(c_cursor)
        print("Here is your randomized pizza!")
        print("Topping(s):")
        print(cursor[0:n])
        print()
        print("Sauce:")
        print(s_cursor[1])
        print()
        print("Crust:")
        print(c_cursor[1])
```

```python
def get_amount(f):
    while True:
        amount = input("How many toppings would you like? (Standard cheese doesn't count): ")
        try:
            val = int(amount)
            if val > f:
                print("Too much, try again")
            elif val < 0:
                print("No negatives buddy")
            elif val == 0:
                print("No toppings? Why are you here?")
            else:
                break
        except ValueError:
            print("Amount must be a number, try again")
    print()
    return val
```

```python
p = str(input("Do you already have an account? Please type YES or NO. "))
if p == 'YES':
    p = old_person()
    p.login()
elif p == 'NO':
    p = new_person()
    p.sign_up()
```
